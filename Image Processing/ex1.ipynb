```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
N = 200
x = np.linspace(-2,2,N)
y = x

X,Y = np.meshgrid(x,y)
R = np.sqrt(X**2 + Y**2)
A = np.zeros(np.shape(R))
A[np.where(R<0.5)] = 1.0



from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

circle = ax.plot_surface(X,Y,A,cmap = 'viridis')
fig.colorbar(circle)

```
